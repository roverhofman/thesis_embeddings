```python
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import GramianAngularField
import torch
from torchvision import models, transforms
from PIL import Image

# 1) Load your splits
df_train = catalog.load("windows_90_train")
df_val   = catalog.load("windows_90_val")
df_test  = catalog.load("windows_90_test")

# 2) Assume each DataFrame row is a window of length 90
X_train = df_train.values
X_val   = df_val.values
X_test  = df_test.values

# 3) Convert time series windows to Gramian Angular Field images
image_size = 90
gaf = GramianAngularField(image_size=image_size, method='summation')

images_train = gaf.fit_transform(X_train)   # shape: (n_train, 64, 64)
images_val   = gaf.fit_transform(X_val)         # shape: (n_val,   64, 64)
images_test  = gaf.fit_transform(X_test)        # shape: (n_test,  64, 64)

# 4) Helper to display a few samples
def show_samples(images, n=15, title=''):
    plt.figure(figsize=(10, 2))
    for i in range(n):
        plt.subplot(1, n, i+1)
        plt.imshow(images[i], cmap='rainbow', origin='lower')
        plt.axis('off')
    plt.suptitle(title, y=1.15)
    plt.tight_layout()
    plt.show()

# 5) Display sample images from each split
show_samples(images_train, title='Train GAF Samples')
show_samples(images_val,   title='Validation GAF Samples')
show_samples(images_test,  title='Test GAF Samples')

# 6) (Optional) Define CNN-based embedding extractor
backbone = models.resnet18(pretrained=True)
feature_extractor = torch.nn.Sequential(*list(backbone.children())[:-1])
feature_extractor.eval()

preprocess = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485], std=[0.229]),
])

def get_ts_image_embedding(img: np.ndarray, device='cpu'):
    # img: H×W array in arbitrary range → normalize to [0,1]
    arr = (img - img.min()) / (img.ptp() + 1e-8)
    pil = Image.fromarray((255 * arr).astype('uint8'))
    x = preprocess(pil).unsqueeze(0).to(device)
    with torch.no_grad():
        feats = feature_extractor(x)
    feats = feats.flatten(1)
    feats = feats / feats.norm(dim=1, keepdim=True)
    return feats.cpu().squeeze(0).numpy()

# 7) Example: compute embedding for first training image
embedding_example = get_ts_image_embedding(images_train[0])
print('Embedding vector shape:', embedding_example.shape)
```
