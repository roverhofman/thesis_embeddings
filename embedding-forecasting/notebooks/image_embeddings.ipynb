```python
import numpy as np
import pandas as pd
from pyts.image import GramianAngularField, MarkovTransitionField, RecurrencePlot
from PIL import Image
import torch
from torchvision.models import resnet50, ResNet50_Weights
from torchvision.transforms import Compose, Resize, ToTensor, Normalize
from transformers import CLIPProcessor, CLIPModel, AutoImageProcessor, AutoModel
```

```python
# 1) Load your splits
df_train = catalog.load("windows_90_train")
df_val   = catalog.load("windows_90_val")
df_test  = catalog.load("windows_90_test")
```

```python
# Functions to transform time series to images
def ts_to_gaf(ts, image_size=None):
    ts = np.array(ts).reshape(1, -1)
    gaf = GramianAngularField(image_size=image_size, method='summation')
    img = gaf.transform(ts)[0]
    return img

def ts_to_mtf(ts, image_size=None):
    ts = np.array(ts).reshape(1, -1)
    mtf = MarkovTransitionField(image_size=image_size)
    img = mtf.transform(ts)[0]
    return img

def ts_to_rp(ts, image_size=None):
    ts = np.array(ts).reshape(1, -1)
    rp = RecurrencePlot()
    img = rp.transform(ts)[0]
    return img

def ts_to_image(ts, method='gaf', image_size=None):
    # if no size specified, use series length
    L = len(ts)
    if image_size is None:
        image_size = L
    # pick transform
    if method == 'gaf':
        img = ts_to_gaf(ts, image_size)
    elif method == 'mtf':
        img = ts_to_mtf(ts, image_size)
    elif method == 'rp':
        img = ts_to_rp(ts, image_size)
    else:
        raise ValueError(f"Unknown method: {method}")
    # normalize to [0,255]
    img_min, img_max = img.min(), img.max()
    if img_max > img_min:
        img = (img - img_min) / (img_max - img_min)
    else:
        img = np.zeros_like(img)
    img = (img * 255).astype(np.uint8)
    # to RGB PIL
    return Image.fromarray(img).convert("RGB")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

clip_model = CLIPModel.from_pretrained(
    "openai/clip-vit-base-patch32",
    device_map="cpu",            # or your desired map
    use_safetensors=True         # <- force safetensors backend
)
clip_model = clip_model.to(device)
clip_model.eval()

# DINOv2 model (Vision Transformer)
dino_processor = AutoImageProcessor.from_pretrained("facebook/dinov2-base")
dino_model = AutoModel.from_pretrained(
    "facebook/dinov2-base",
    device_map="cpu",
    use_safetensors=True
).to(device)
dino_model.eval()

# ResNet50 model (features only)
resnet_weights = ResNet50_Weights.IMAGENET1K_V2
resnet_base = resnet50(weights=resnet_weights)
# Remove final classification layer
resnet_model = torch.nn.Sequential(*list(resnet_base.children())[:-1]).to(device)
resnet_model.eval()
# Preprocessing transform for ResNet
resnet_transform = resnet_weights.transforms()

# Embedding functions for each model
def embed_clip(pil_img):
    inputs = clip_processor(images=pil_img, return_tensors="pt").to(device)
    with torch.no_grad():
        features = clip_model.get_image_features(**inputs)
    return features[0].cpu().numpy()

def embed_dino(pil_img):
    inputs = dino_processor(images=pil_img, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = dino_model(**inputs)
    # Use the last hidden state of [CLS] token as image embedding
    last_hidden = outputs.last_hidden_state  # shape [1, seq_len, hidden_dim]
    embedding = last_hidden[0, 0, :].cpu().numpy()
    return embedding

def embed_resnet(pil_img):
    tensor = resnet_transform(pil_img).unsqueeze(0).to(device)
    with torch.no_grad():
        feat = resnet_model(tensor)
    # Flatten output (shape [1, 2048, 1, 1] -> [2048])
    embedding = feat.squeeze().cpu().numpy()
    return embedding

# Function to process a DataFrame and save embeddings
def process_and_save(df, name, methods=['gaf','mtf','rp']):
    """
    Processes a DataFrame of time series windows and saves embeddings.
    Each row in df is expected to contain the time series window as the first element.
    """
    for method in methods:
        # Generate images for each window in the dataset
        images = []
        for _, row in df.iterrows():
            ts = row.iloc[0]
            pil_img = ts_to_image(ts, method=method)
            images.append(pil_img)
        # Apply each embedding model
        clip_embs, dino_embs, resnet_embs = [], [], []
        for img in images:
            clip_embs.append(embed_clip(img))
            dino_embs.append(embed_dino(img))
            resnet_embs.append(embed_resnet(img))
        # Convert to numpy arrays
        clip_embs = np.vstack(clip_embs)
        dino_embs = np.vstack(dino_embs)
        resnet_embs = np.vstack(resnet_embs)
        # Save embeddings to disk
        np.save(f"{name}_{method}_clip.npy", clip_embs)
        np.save(f"{name}_{method}_dino.npy", dino_embs)
        np.save(f"{name}_{method}_resnet.npy", resnet_embs)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Load your splits
df_train = catalog.load("windows_90_train")
df_val   = catalog.load("windows_90_val")
df_test  = catalog.load("windows_90_test")

# 2) Take first 3 train windows
sample_df = df_train.head(3)
methods   = ['gaf', 'mtf', 'rp']

# 3) Plot each sample’s images
for idx, (_, row) in enumerate(sample_df.iterrows()):
    ts = row.to_numpy()   # ← now ts is the full window array, not a single float
    plt.figure(figsize=(12, 4))
    for j, method in enumerate(methods):
        img = ts_to_image(ts, method=method)    # auto‐sizes to len(ts)
        plt.subplot(1, 3, j+1)
        plt.imshow(img)
        plt.title(method.upper())
        plt.axis('off')
    plt.suptitle(f"Train sample #{idx}")
    plt.show()

# 4) Compute & print embedding shapes for those samples
for method in methods:
    print(f"\n=== Embeddings for {method.upper()} ===")
    clip_embs, dino_embs, resnet_embs = [], [], []
    for _, row in sample_df.iterrows():
        ts = row.to_numpy()                  # full array
        img = ts_to_image(ts, method=method)
        clip_embs.append(embed_clip(img))
        dino_embs.append(embed_dino(img))
        resnet_embs.append(embed_resnet(img))
    print("CLIP:   ",   np.vstack(clip_embs).shape)
    print("DINO:   ",   np.vstack(dino_embs).shape)
    print("ResNet: ",   np.vstack(resnet_embs).shape)
```
